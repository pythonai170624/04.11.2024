```python
nums = np.arange(0, 20)
print(nums[(nums > 4) & (nums < 20)])
print(nums[(nums % 2 == 0) | (nums % 3 == 0)])
```

```python
np_array = np.arange(0, 25).reshape(5, 5)
print(np_array)
print('sum no axis', np_array.sum(axis=None))
print('sum axis=0 columns:\n', np_array.sum(axis=0))
print('sum axis=1 rows:   \n', np_array.sum(axis=1))
print('0 + 5 + 10 + 15 + 20 = ', 0 + 5 + 10 + 15 + 20)
print('col 1', np_array[:, 1].sum())  # sum col
print('row 1', np_array[1, :].sum())  # sum ROW 1
```

```python
np.random.seed(101)
rand_arr = np.random.randint(1, 101, size=(5, 5))
print(rand_arr)

print('sum rows:', np.sum(rand_arr, axis=1))
#print(96 + 88 + 5 + 65 + 50)
print('sum rows:', np.min(rand_arr, axis=1))
print('sum rows:', np.max(rand_arr, axis=1))
print('sum rows:', np.mean(rand_arr, axis=1))
print('sum rows:', np.median(rand_arr, axis=1))

print(rand_arr[rand_arr > 10])
```
